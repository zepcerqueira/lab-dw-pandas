```python
#Testing for income:

income_mean = insurance_df["Income"].mean()
income_median = insurance_df["Income"].median()
income_mode = insurance_df["Income"].mode()
income_std = insurance_df["Income"].std()
print("Income Mean - ", round(income_mean, 3))
print("Income Median - ", round(income_median, 3))
print("Income Mode - ", round(income_mode, 3))
print("Income Stand. Deviat. - ", round(income_std, 3))

#Median is more reliable because it takes out extreme values, like zero and max 99960.
#mode = 0 as the most freq values indicates that we have a lot of data in income as zero.
```

```python
#Challenge 2: analyzing the data
#Exercise 1
#The marketing team wants to know the top 5 less common customer locations. 
#Create a pandas Series object that contains the customer locations and their frequencies, 
#and then retrieve the top 5 less common locations in ascending order.

import numpy as np
import pandas as pd

data_url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"

insurance_mark_df = pd.read_csv(data_url)
#insurance_mark_df

less_common_location = insurance_mark_df["ST"].value_counts()
print(less_common_location.sort_values(ascending=True).head(1))
less_common_location
```

```python
#Exercise 2
#The sales team wants to know the total number of policies sold for each type of policy. 
#Create a pandas Series object that contains the policy types and their total number of policies sold, 
#and then retrieve the policy type with the highest number of policies sold.

import numpy as np
import pandas as pd

data_url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"

insurance_mark2_df = pd.read_csv(data_url)
#insurance_mark_df

number_policies = insurance_mark2_df["Policy Type"].value_counts()
number_policies

max_sold_policy = insurance_mark2_df['Policy Type'].value_counts().idxmax()
max_sold_policy 
```

```python
#Exercise 3
#The sales team wants to know if customers with Personal Auto have 
#a income than those with Corporate Auto. How does the average income 
#compare between the two policy types?

import numpy as np
import pandas as pd
data_url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
insurance_mark3_df = pd.read_csv(data_url)

#Get the highest income of personal auto.

highest_income_pers_auto = insurance_mark3_df.groupby("Policy Type")["Income"].median()
highest_income_pers_auto

#Clients with special auto have a higher income. Next is corporate auto.
```
